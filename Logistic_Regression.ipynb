From the table above we can find out which one is dependent variables and which one is independent variables. Lets say x is dependent variable and y is independent variables. Column diagnosis is dependent and remaining columns are indedependent variables. Now lets split into x and y.


```python
x=df.iloc[:,2:].values
y=df.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
train_accs , test_accs=[],[]
#Now test the data set. For now I will test 25000 datasets
for i in range(25000):
    X_train,X_test,Y_train,Y_test=train_test_split (x,y,test_size=0.25,random_state=i)


    #Scaling the features
    scalar=StandardScaler()
    X_train=scalar.fit_transform(X_train)
    X_test=scalar.fit_transform(X_test)

    lrc=LogisticRegression()

    lrc.fit(X_train,Y_train)

    y_pred_train=lrc.predict(X_train)
    y_pred_test=lrc.predict(X_test)
```

```python
y_pred_train
```

```python
y_pred_test
```

```python
  # Compute the accuracy of the classifier on the training and testing data
train_accuracy = accuracy_score(Y_train, y_pred_train)
test_accuracy = accuracy_score(Y_test, y_pred_test)
train_accs.append(train_accuracy)
test_accs.append(test_accuracy)
```

```python
#Compute the average of training and testing accuracies

train_acc_mean=np.mean(train_accs)
test_acc_mean=np.mean(test_accs)
# Print the average training and testing accuracies
print("Average Train Accuracy: {:.2f}%".format(train_acc_mean * 100))
print("Average Test Accuracy: {:.2f}%".format(test_acc_mean * 100))
```
